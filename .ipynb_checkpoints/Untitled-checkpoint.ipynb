```python
import pandas as pd
```

```python
df = pd.read_csv("adult.csv")
```

```python
df
```

```python
df.workclass.value_counts()
```

```python
df.occupation.value_counts()
```

```python
df = pd.concat((df.drop('occupation', axis=1), pd.get_dummies(df.occupation).add_prefix("occupation_")), axis=1)
df = pd.concat((df.drop('workclass', axis=1), pd.get_dummies(df.workclass).add_prefix("workclass_")), axis=1)
df = pd.concat((df.drop('marital-status', axis=1), pd.get_dummies(df['marital-status']).add_prefix("marital-status_")), axis=1)
df = pd.concat((df.drop('race', axis=1), pd.get_dummies(df.race).add_prefix("race_")), axis=1)
df = pd.concat((df.drop('relationship', axis=1), pd.get_dummies(df.relationship).add_prefix("relatonship_")), axis=1)
df = df.drop('education', axis=1)
df = pd.concat((df.drop('native-country', axis=1), pd.get_dummies(df['native-country']).add_prefix("native-country_")), axis=1)
```

```python
df
```

```python
df['gender'] = df['gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['income'] = df['income'].apply(lambda x: 1 if x == '>50k' else 0)
```

```python
df
```

A correlation heat map for the corelation of the features

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 12))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
```

```python
df.corr()
```

```python
corelations = df.corr()['income'].abs()
sorted_corelations = corelations.sort_values()
num_cols_to_drop = int(0.8 * len(df.columns))
cols_to_drop = sorted_corelations.iloc[:num_cols_to_drop].index
df_dropped = df.drop(cols_to_drop, axis=1)
```

```python
df_dropped
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df_dropped.corr(), annot=True, cmap='coolwarm')
```

Training the data after splitting it

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
train_df
```

```python
test_df
```

```python
train_x = train_df.drop('income', axis=1)
train_y = train_df['income']

test_x = test_df.drop('income', axis=1)
test_y = test_df['income']
```

```python
forest = RandomForestClassifier()

forest.fit(train_x, train_y)
```

```python
forest.score(test_x, test_y)
```
